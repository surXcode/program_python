```python
# 4. WAJP to print sum of numbers from 1 to 100 by using recursion.


def sum_num(n):
    if n == 0:
        return 0
    else:
        return n + sum_num(n - 1)


result = sum_num(100)
print("sum of numbers 1 to 100 =",result)
```

```python
# 5. WAJP to print sum of squares of numbers from 1 to 100 by using recursion.


def sum_of_square(n):
    if n == 0:
        return 0
    else:
        return n * n + sum_of_square(n - 1)   
result = sum_of_square(100)
print("Sum of square numbers from 1 to 100 =", result)
```

```python
# 6. WAJP to print sum of cubes of numbers from 1 to 100 by using recursion.as_integer_ratio.

def sum_of_cube(n):
    if n == 0:
        return 0
    else:
        return n*n*n + sum_of_cube(n - 1)
result = sum_of_cube(5)
print("Sum of cube numbers from 1 to 5 =", result)
```

```python
# 7. WAJP to calculate factorial by using recursion.


def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))
```

```python
# 8. WAJP to calculate power by using recursion.

def power(base, exp):
    if exp == 0:          
        return 1
    else:
        return base * power(base, exp - 1)   
print(power(6, 3)) 
```

```python
# 16. Given an integer n, return true if it is a power of two. Otherwise, return false.

An integer n is a power of two, if there exists an integer x such that n == 2x.



class Solution:
    def isPowerOfTwo(self, n):
        if n == 1:
            return True
        if n < 1 or n % 2 != 0:
            return False
        return self.isPowerOfTwo(n // 2)
```
