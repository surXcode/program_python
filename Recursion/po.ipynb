* It follows start mechanism.

# Base case:

* A condition in a recursive function that returns a value without making another recursive call.

* Stops recursion.

# Recursive case:

* The condition or code in a recursive function that reduces the problem and calls the same function again.


# ðŸ“¦ DIFFERENCE BETWEEN BASE CASE AND RECURSIVE CASE: 

| S.No |        BASE CASE                           |         RECURSIVE CASE                          |
|------|---------------------------------------------|------------------------------------------------|
|  1.  | It is the condition that stops recursion.   | It is the part that calls the function itself.  |
|  2.  | Gives a direct solution.                    | Breaks the problem into smaller parts.          |
|  3.  | Executes only once when condition is met.   | Executes repeatedly until base case is reached. |
|  4.  | Does not make any further function calls.   | Makes a call to the same function again.        |
|  5.  | Prevents infinite recursion.                | Keeps recursion going.                          |
|  6.  | Example: if n == 0: return 1                | Example: return n * factorial(n - 1)    

```python
def main(n):
    print(n)
    main(n+1)
main(1)



```

```python
# Print 1 to 10 natural numbers.

def numbers(n=10):
    if n==1:           # Base case
        print(n)
    else:
        numbers(n-1)    # Recursive case
        print(n)

numbers()
```

```python
def numbers(n=10):
    if n==1:           # Base case
        print(n)
    else:
        print(n)
        numbers(n-1)    # Recursive case
numbers()
```

```python
# Factorial of a number.

def factorial(n):
    if n==1 or n==0:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(4))
print(factorial(5))
print(factorial(0))
```
